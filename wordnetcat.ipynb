```python
def getsets(word):
    synverb =set()
    antoverb = set()
    category = 'v'
    synsetsForWord = wordnet.synsets(word, pos=category)
    syall = wordnet.synsets(word)
    for syn in synsetsForWord:
        for l in syn.lemmas():
            synverb.append(l.name())
            if l.antonyms():
                antoverb.append(l.antonyms()[0].name())
    return synverb
```
